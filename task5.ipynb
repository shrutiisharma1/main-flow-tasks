```python
# Create age groups (bins)
age_bins = pd.cut(df['age'], bins=np.arange(30, 81, 10))

# Count the number of heart disease cases by age group
heart_disease_by_age_group = df[df['target'] == 1].groupby(age_bins)['target'].count()
```

```python
# Plotting the bar plot
plt.figure(figsize=(10, 6))
heart_disease_by_age_group.plot(kind='bar', color='teal')
plt.title('Number of Heart Disease Cases by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Number of Heart Disease Cases')
```
